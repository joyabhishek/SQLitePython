```python
import pandas as pd
import numpy as np
from sqlite3 import dbapi2 as sq3
import os
```

```python
createTableSchema = '''

    DROP TABLE IF EXISTS cardioGoodFitness;

    CREATE TABLE "cardioGoodFitness" (
        "id" INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        "Product" VARCHAR NOT NULL,
        "Age" INTEGER,
        "Gender" VARCHAR,
        "Education" INTEGER,
        "MaritalStatus" VARCHAR,
        "Usage" VARCHAR,
        "Fitness" INTEGER,
        "Income" INTEGER,
        "Miles" INTEGER
    );
'''
```

```python
def get_db(dbfile):
    sqllite_db = sq3.connect(os.path.join(".", dbfile))
    return sqllite_db
```

```python
def init_db(db_file, schema):
    db = get_db(db_file)
    db.executescript(schema)
    db.commit()
    return db
```

```python
df = pd.read_csv('CardioGoodFitness.csv')
df.head()
```

```python
db = init_db('cardioGoodFitness.db', createTableSchema)
```

```python
df.to_sql('cardioGoodFitness', db, if_exists='append', index=False)
```

```python
sel = """
 SELECT * FROM cardioGoodFitness;
"""
```

```python
db.executescript(sel).fetchall()
```
